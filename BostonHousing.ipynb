```
import pandas as pd

df= pd.read_csv('BostonHousing.csv')
df= df.rename(columns={'medv': 'TARGET','rm':'RM','lstat':'LSTAT'})
```

```
df[["LSTAT","RM","TARGET"]].head()
```

```
plt.figure(figsize=(8,4))
features =df[["LSTAT","RM"]]
target = df["TARGET"]
plt.subplot(1,2,1)
plt.scatter(features["LSTAT"],target)
plt.xlabel("LSTAT")
plt.xlabel("TARGET")
plt.subplot(1,2,2)
plt.scatter (df["LSTAT"], target)
plt.subplot(1,2,2)
plt.scatter(df["RM"], target)
plt.xlabel("RM")
plt.xlabel("TARGET")
plt.show()
```

```
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)
```

```
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
```

```
plt.figure(figsize=(5,4))
plt.scatter(y_test,y_pred)
plt.xlabel("Actual y")
plt.ylabel("Predicted y")
plt.show()
plt.subplot(1,2,1)
```

```
intercept = model.intercept_

coff = model.coef_
```

```
#Evaluate the model 'sperfformance
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score # Import mean_absolute_error

mse = mean_squared_error(y_test, y_pred)
print('Mean Square Error:', mse)

mae = mean_absolute_error(y_test, y_pred) # Now mean_absolute_error is defined and can be used
print("Mean Absolute Error:", mae)

r2 = r2_score(y_test, y_pred)
print('R2 Score:',r2)
```
